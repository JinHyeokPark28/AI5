```python
# 이진트리의 속성
# 트리의 개수 최대 2개
#    루트
# 왼쪽 오른쪽

# 이진트리의 종류
# 포화이진트리
# 완전 이진트리

# 이진트리 구현
# 리스트를 통해서 구현을 합니다.

def BinaryTreeList(root):
    return [root, [], []]

def insertLeft(root, newBranch):
    t= root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])
    return root

# 오른쪽 노드 추가
def insertRight(root, newBranch):
    t= root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch,[], t])
    else:
        root.insert(2, [newBranch, [], []])
    return root
# 루트의 값 가져오기
def GetRootVal(root):
    return root[0]
# 왼쪽자식값 가져오기
def GetLeftChild(root):
    return root[1]
# 오른쪽 자식값 가져오기
def GetRightChild(root):
    return root[2]

def main():
    r = BinaryTreeList(3)
    insertLeft(r, 4)
    insertLeft(r, 5)
    insertRight(r, 6)
    insertRight(r, 7)
    print(GetRootVal(r))
    print(GetLeftChild(r))
    print(GetRightChild(r))

if __name__ == "__main__":
    main()
```
